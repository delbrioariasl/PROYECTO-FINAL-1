# Combinación de los dos

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Define los datos para el radar chart
competiciones = ['Goles', 'Media % Pases', 'Tiros', 'Recuperaciones por partido', 'Pases finales']
datos_eurocopa = [6, 83.6, 77, 67.75, 55]
datos_mundial = [18, 85, 155, 66, 123]

# Ajusta los ángulos del radar chart
angulos = np.linspace(0, 2*np.pi, len(competiciones), endpoint=False).tolist()
angulos += angulos[:1]

# Ajusta los datos para formar un loop cerrado en el gráfico
datos_eurocopa += datos_eurocopa[:1]
datos_mundial += datos_mundial[:1]

# Crea una figura y un subgráfico para el radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plotea los datos de la Eurocopa 2022 en el radar chart
ax.plot(angulos, datos_eurocopa, label='Eurocopa 2022')

# Rellena el área del radar chart de la Eurocopa 2022 con colores
ax.fill(angulos, datos_eurocopa, alpha=0.25)

# Plotea los datos del Mundial 2023 en el radar chart
ax.plot(angulos, datos_mundial, label='Mundial 2023')

# Rellena el área del radar chart del Mundial 2023 con colores
ax.fill(angulos, datos_mundial, alpha=0.25)

# Añade las etiquetas de los datos
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(competiciones)

# Añade una leyenda y un título
ax.legend()
plt.title('Comparación rendimiento en la Eurocopa frente al Mundial')

# Verifica si la carpeta de destino existe, si no, la crea
ruta_destino = r'C:\Users\delbr\Desktop\Proyectos IronHack\PROYECTO_FINAL\imagenes-videos\imagenes'
if not os.path.exists(ruta_destino):
    os.makedirs(ruta_destino)

# Guarda la gráfica como una imagen en la ruta de destino
ruta_guardado = os.path.join(ruta_destino, 'radar_chart.png')
plt.savefig(ruta_guardado, dpi=300)

# Muestra el radar chart comparativo
plt.show()
```
